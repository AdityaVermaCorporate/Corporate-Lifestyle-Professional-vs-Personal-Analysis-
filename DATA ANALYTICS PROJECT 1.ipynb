```python
# CORPORATE WORK LIFESTYLE ANALYSIS

import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np 


data = pd.read_csv("/Users/adityaverma/Downloads/Impact_of_Remote_Work_on_Mental_Health.csv")
df = pd.DataFrame(data)
print(df)

# Checking the integrity of the dataset:
df.info()
df.describe()
# checking for null values :
df.isnull().sum()
# Handling the missing values:
Cleaned = df.dropna(subset= ["Mental_Health_Condition" , "Physical_Activity"])
Cleaned_df = Cleaned.groupby("Industry")["Hours_Worked_Per_Week"].mean().sort_values()
Clean = pd.DataFrame(Cleaned_df)
print(type(Clean))
```

```python
sns.pairplot(data = Cleaned , height = 2.5 ,
             aspect = 2 ,
             diag_kind = "hist")
             
            
plt.show()
```

```python
order = ["Manufacturing" , "Healthcare" , "IT" , "Consulting" , "Education" , "Retail" , "Finance"]
sns.barplot(data = Clean ,
            x = "Hours_Worked_Per_Week" ,
            y = "Industry" ,
            order = order ,
            saturation = 0.70 ,
            estimator = "mean",
        )
plt.title("Manufacturing industry with the highest working hours per week" , size = 15)
plt.show()
```

```python
sns.violinplot(data = Cleaned, 
               x = "Stress_Level",
               y = "Hours_Worked_Per_Week" , hue = "Gender")
plt.title("Males have the highest stress levels" , size  = 15)
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

```python
sns.boxplot(data = Cleaned , x = "Age" , 
            hue = "Mental_Health_Condition")
plt.legend(bbox_to_anchor  = (1.3,1))
plt.title("Mental Health By Age Groups" , size  = 15)
plt.show()
```

```python
sns.barplot(data = Cleaned , x = "Productivity_Change" , hue = "Physical_Activity" )
plt.legend(bbox_to_anchor = (1.3,1))
plt.title("Productivity Increaes with Increase in Physical Exercise" , size  = 15)

```

```python
sns.boxplot(data = Cleaned , x = "Job_Role" ,
            y = "Work_Life_Balance_Rating" ,
            hue = "Region"  , 
            saturation = 1)
plt.legend(bbox_to_anchor = (1.3,1))
plt.title("Productivity Increaes with Increase in Physical Exercise" , size  = 15)
```
