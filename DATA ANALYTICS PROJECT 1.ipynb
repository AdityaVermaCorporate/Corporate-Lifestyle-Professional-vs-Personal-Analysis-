```python
# CORPORATE WORK LIFESTYLE ANALYSIS

import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

data = pd.read_csv("/Users/adityaverma/Downloads/Impact_of_Remote_Work_on_Mental_Health.csv")
df = pd.DataFrame(data)
print(df)
# CHecking for null values //
df.info()
#statistical measure //
df.describe()
df["Physical_Activity"].isnull().agg(sum) # missing values : 1629
df["Mental_Health_Condition"].isnull().agg(sum) # missing values : 1196
Clean_PE = (df["Physical_Activity"] == "Weekly").sum()
print(Clean_PE)
# Handling Missing Values 
df.dropna(inplace = True)
df.describe()

# Moving onto Exploratory Data Analysis :
sns.pairplot(data = df , diag_kind  = "kde")
plt.savefig("PAIRPLOT.png" , dpi =300 ,pad_inches = 0.2)
plt.show()
# Saving the file.csv to computer
df.to_csv("CORPORATE JOBS ANALYSIS.csv" , index  = False)





```
